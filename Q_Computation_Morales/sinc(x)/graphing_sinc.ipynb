```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation
```

```python
#vars
t = np.arange(-10, 10.1, 0.1)
wfi = 1
w = np.arange(0, 2.01, 0.01)

#meshgrid
T, W = np.meshgrid(t, w)

#stw
STW = (np.sin((wfi - W) * T / 2) / ((wfi - W) / 2))**2

#print sizes like prof, don't need this personally
#print("Size of STW:", STW.shape)
#print("Size of t:", t.shape)
#print("Size of w:", w.shape)
#print("Size of T:", T.shape)
#print("Size of W:", W.shape)

#plot mesh
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(T, W, STW, cmap='viridis')
ax.set_xlabel('t')
ax.set_ylabel(r'$\omega$')
ax.set_zlabel(r'$\sin^2((\omega_{fi}-\omega)t/2)/((\omega_{fi}-\omega)/2)^2,\quad \omega_{fi}=1$')

plt.savefig('sinc_w_python.png')
plt.show()
```

```python
#plot element
surface = ax.plot_surface(T, W, STW, cmap='viridis')

#func for new angle
def update(frame):
    ax.view_init(elev=30, azim=frame)
    return [surface]

#ani call
ani = animation.FuncAnimation(fig, update, frames=np.arange(0, 360, 1), blit=True)

#save gif
ani.save('spinning_sinc.gif', writer='imagemagick', fps=30)



```
