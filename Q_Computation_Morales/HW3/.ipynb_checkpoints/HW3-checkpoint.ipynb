### Problem 1:
- Similar (same?) to problem we did in class 2/20. 
- Gameplan
    - Argue ring of length L w boundry conditions forces psi(x) as shown
    - Show thoes energy satisfy boundry conditions of unperturbed energies
    - We have matrix elements H', can use integral to pick out transitions n --> m.
    - Look at subspace spanned by Psi**0_(+/- 1), show diagonal and off diagonal elements are zero

```python
#symbols
L, x, m, n, hbar, mass = sp.symbols('L x m n hbar m', real=True)
v0 = sp.Symbol('v0', real=True)


#part a, unperturbed eigenstates
k_n = (2 * sp.pi * n) / L
psi_n = (1/sp.sqrt(L)) * sp.exp(1j * k_n * x)
print('k_n =')
display(k_n)
print('psi_n = ')
display(psi_n)
E_n = (hbar**2 * k_n**2) / (2 * mass) #unperturbed energy eigenvalues
print('unperturbed energy eigenvalues: ')
display(E_n)

#part b, matrix element of h'_mn
H_prime = v0 * sp.cos(2 * sp.pi * x / L)
print('H_prime_mn =')
display(H_prime)

#compute matrix element h'_mn = ⟨ψ_m | H' | ψ_n⟩ --> explicit sub bc sympy weird
h_mn = sp.integrate(
    sp.conjugate(psi_n).subs(n, m) * H_prime * psi_n.subs(n, n), (x, 0, L)
).simplify()

#use kronecker delta explicitly for simplification
delta_expr = (v0/2) * (sp.KroneckerDelta(m, n+1) + sp.KroneckerDelta(m, n-1))

# Display results
h_mn_simplified = h_mn.simplify()
h_mn_explicit = h_mn_simplified.subs(h_mn_simplified, delta_expr)
print('H_mn after kroneckerdelta sub is: ')
display(h_mn_explicit)
#now part c construct matrix in subspace n = ±1
W_matrix = sp.Matrix([
    [h_mn.subs({m: 1, n: 1}), h_mn.subs({m: 1, n: -1})],
    [h_mn.subs({m: -1, n: 1}), h_mn.subs({m: -1, n: -1})]
])

#compute and display eigenvalues for first order energy correc
W_eigenvalues = W_matrix.eigenvals()

print('the matrix W in subspace n = ±1 ')
display(W_matrix)
print('the eigenvalues for the first order energy corrections are:')
W_eigenvalues_df = pd.DataFrame(
    list(W_eigenvalues.items()), columns=['Eigenvalue', 'Multiplicity']
)
display(W_eigenvalues_df)


```

### Problem 2:


```python
#symbols
Z, alpha, m, c, n, l, j = sp.symbols('Z alpha m c n l j', real=True, positive=True)

#ke relativistic corrections
E_rel = - (Z * alpha)**4 * m * c**2 / (2 * n**4) * (4/n - 3/(l + sp.Rational(1/2)))

#spin orbit correction
E_so = (Z * alpha)**4 * m * c**2 / (n**3) * (l / ((l + sp.Rational(1/2)) * (l + 1)))

#fine struct. corrections
E_fs = E_rel + E_so

#show in terms of j
E_final = E_fs.subs(j, l - sp.Rational(1/2))

display(E_rel, E_so,E_final)
```

## Problem 3:

```python
#we use/define spherical
theta, phi = sp.symbols('theta phi')
ax, ay, az, bx, by, bz = sp.symbols('a_x a_y a_z b_x b_y b_z')

#def rhat in sphere (ugh lol)
r_hat = sp.Matrix([sp.sin(theta) * sp.cos(phi), sp.sin(theta) * sp.sin(phi), sp.cos(theta)])
print('r_hat is:')
display(r_hat)


#def vectors a and b
a_vec = sp.Matrix([ax, ay, az])
b_vec = sp.Matrix([bx, by, bz])
print('vects a, and b are: ')
display(a_vec,b_vec)

#dot each w/ rhat
a_dot_r = a_vec.dot(r_hat)
b_dot_r = b_vec.dot(r_hat)
display('a and b dotted with r is: ')
display(a_dot_r,b_dot_r)

#set up that there integrand
integrand = (a_dot_r * b_dot_r) * sp.sin(theta)
print('the super long integrand that I had to do by hand :( :')
display(integrand)

#integrateee
#over theta and phi
integral = sp.integrate(integrand, (phi, 0, 2*sp.pi), (theta, 0, sp.pi))

#simplify it
integral_simplified = sp.simplify(integral)

print('after integrating:')
display(integral_simplified)
```

```python
#for second part S and r def
Se_x, Se_y, Se_z = sp.symbols('Se_x Se_y Se_z')
r_x, r_y, r_z = sp.symbols('r_x r_y r_z')
r = sp.symbols('r', positive=True)

#def spin vect
S_e = sp.Matrix([Se_x, Se_y, Se_z])
print('our spin vector S_e is:')
display(S_e)

#S_e dot r
S_e_dot_r = S_e.dot(r_hat)
print('the dot of S_e and r is: ')
display(S_e_dot_r)

#compute full term of expr given by prof: (3(S_e . r)(S_e . r_hat) - S_e . S_e) / r**3
expr = (3 * (S_e_dot_r * S_e_dot_r) - S_e.dot(S_e)) / r**3
print('full expression is: ')
display(expr)

#sub delta identiy w/matrix knowing ⟨r_i r_j⟩ = (1/3) δ_ij to simplify
delta_ij = sp.eye(3) / 3  #1/3 for K.delta
print('our kroniker delta is of the form:')
display(delta_ij)

#loop through ⟨S_e . r S_e . r_hat⟩ 
expected_value = 0
results = []  # To store the results

#loop through i and j
for i in range(3):
    for j in range(3):
        expected_value += 3 * S_e[i] * S_e[j] * delta_ij[i, j]
        results.append({
            'i': i,
            'j': j,
            'expected_value': expected_value
        })

#convert to df for formatting
df = pd.DataFrame(results, columns=['i', 'j', 'expected_value'])

#display df
display("Expected Values Table:", df)

#final expecattion value
expected_value -= S_e.dot(S_e)

#show final
print('final expcatation value: ')
display(expected_value.simplify())
```

### analysis of plots
- ring potential gif
    -unit circle / ring, color changes according to perturb potential V(theta) = lamda v_0 cos(theta), red dot moves to show phase of the n=1 eigenstate
    - we can see as the perturbation (scaled by lamda) turns on, potential along ring is visible 
- energy correction plot
    - horizontal line at 0, shows first-order energy correction for n=+/-1

```python
#plot matrix elements function
def plot_matrix_elements():
    L_num = 1.0 #length of ring
    v0_num = 0.5 #perturb strength
    x_vals = np.linspace(0, L_num, 500)

    #create matrix for m, n indices
    n_values = np.array([-2, -1, 0, 1, 2]) #n range
    m_values = np.array([-2, -1, 0, 1, 2]) #m range
    h_mn_matrix = np.zeros((len(m_values), len(n_values)))

    #nonzero where where m=n+1 or m=n-1
    for i, m in enumerate(m_values):
        for j, n in enumerate(n_values):
            if m == n + 1 or m == n - 1:
                h_mn_matrix[i, j] = v0_num / 2 #here is the non-zero val

    #heatmap
    plt.figure(figsize=(6, 5))
    plt.imshow(h_mn_matrix, cmap='Blues', origin='lower', 
               extent=[n_values[0] - 0.5, n_values[-1] + 0.5, 
                       m_values[0] - 0.5, m_values[-1] + 0.5])
    plt.xticks(n_values)
    plt.yticks(m_values)
    plt.xlabel(r"Index $n$", fontsize=12)
    plt.ylabel(r"Index $m$", fontsize=12)
    plt.title("Matrix Elements $h'_{mn}$", fontsize=14)

    #annotate
    for i in range(len(m_values)):
        for j in range(len(n_values)):
            plt.text(n_values[j], m_values[i], f"{h_mn_matrix[i, j]:.1f}",
                     ha='center', va='center', color='black')

    plt.colorbar(label="Value of $h'_{mn}$")
    plt.tight_layout()
    plt.savefig('p1_h_mn.png',dpi=600)
    plt.show()


#e-state visual funct.
def plot_eigenstate():
    L_val = 1.0 #length
    v0_val = 1.0 #perturb
    x_vals = np.linspace(0, L_val, 400)

    #define eigenstate function for given n
    psi = lambda n, x: (1 / np.sqrt(L_val)) * np.exp(1j * (2 * np.pi * n * x / L_val))
    psi1 = psi(1, x_vals) #estate for n=1
    psi_neg1 = psi(-1, x_vals) #estate for n=-1

    #re and im parts for n=1 and n=-1
    fig, axs = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
    axs[0].plot(x_vals, psi1.real, label='re(psi, n=1)')
    axs[0].plot(x_vals, psi1.imag, label='im(psi, n=1)')
    axs[0].set_title('Eigenstate for n=1')
    axs[0].legend()

    axs[1].plot(x_vals, psi_neg1.real, label='re(psi, n=-1)')
    axs[1].plot(x_vals, psi_neg1.imag, label='im(psi, n=-1)')
    axs[1].set_title('Eigenstate for n=-1')
    axs[1].legend()

    plt.xlabel('x')
    plt.tight_layout()
    plt.savefig('p1_eigenstates.png',dpi=600)
    plt.show()


#heatmap
def plot_h_matrix():
    v0_val = 1.0
    m_range = np.arange(-5, 6)
    n_range = np.arange(-5, 6)
    h_matrix = np.zeros((len(m_range), len(n_range)))

    #populate matrix with nonzero values where m=n+1 or m=n-1
    for i, m_val in enumerate(m_range):
        for j, n_val in enumerate(n_range):
            if m_val == n_val + 1 or m_val == n_val - 1:
                h_matrix[i, j] = v0_val / 2

    #plot the heatmap
    plt.figure(figsize=(6, 5))
    plt.imshow(h_matrix, cmap='viridis', extent=[n_range[0] - 0.5, n_range[-1] + 0.5, 
                                                 m_range[-1] + 0.5, m_range[0] - 0.5])
    plt.colorbar(label='Matrix Element Value')
    plt.xlabel('n')
    plt.ylabel('m')
    plt.title('Heatmap of $h\'_{mn}$')
    plt.savefig('p1_H_heatmap.png',dpi=600)
    plt.show()


#animate the perturbation potential on a ring function
def animate_ring():
    L_val = 2 * np.pi #circumfrence
    v0_val = 1.0 #strength
    mass = 1.0 #mass arbitrary
    hbar = 1.0 #unitless/atomic
    E1 = (1**2) / 2.0 #energy for n=1

    theta = np.linspace(0, 2 * np.pi, 200) #ang coords
    ring_x = np.cos(theta)
    ring_y = np.sin(theta) #y coord,  x above

    points = np.array([ring_x, ring_y]).T.reshape(-1, 1, 2) #np array point on ring
    segments = np.concatenate([points[:-1], points[1:]], axis=1) #connect points

    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_aspect('equal')
    ax.set_xlim([-1.2, 1.2])
    ax.set_ylim([-1.2, 1.2])
    ax.set_title('Ring with Perturbation Potential & Eigenstate Phase', fontsize=12)
    ax.set_xlabel('x')
    ax.set_ylabel('y')

    lam = 0.0
    potential = lam * v0_val * np.cos(theta) #potentail along ring
    norm = plt.Normalize(vmin=-v0_val, vmax=v0_val) #normalize
    lc = LineCollection(segments, cmap='viridis', norm=norm) #line collection for ring
    lc.set_array(potential) 
    lc.set_linewidth(3)
    ax.add_collection(lc) 

    cbar = plt.colorbar(lc, ax=ax, pad=0.1) #colorbar
    cbar.set_label('v0 * cos(theta)') 

    marker, = ax.plot([], [], 'ro', markersize=8) #estate marker
    text_lam = ax.text(-1.1, 1.1, '', fontsize=10, bbox=dict(facecolor='white', alpha=0.6)) #this adds the lambda text

    frames = 100 #total frames
    dt = 0.1 #timestep in phase evo

    def update(frame): 
        lam = frame / (frames - 1) #lamda param
        potential = lam * v0_val * np.cos(theta) #update potential
        lc.set_array(potential) 

        t = frame * dt
        phi = (-E1 * t) % (2 * np.pi) #calc phase
        marker.set_data([np.cos(phi)], [np.sin(phi)])

        text_lam.set_text(f'lambda = {lam:.2f}')
        return lc, marker, text_lam 

    ani = animation.FuncAnimation(fig, update, frames=frames, interval=50, blit=True)
    ani.save('p1_animated_ring.gif', writer='pillow', fps=20)
    plt.tight_layout()
    plt.show()


#funct calls
plot_matrix_elements()
plot_eigenstate()
plot_h_matrix()
animate_ring()
```

```python
#symbols
Z, alpha, m, c, n, l, j = sp.symbols('Z alpha m c n l j', real=True, positive=True)

#rel energy correction
E_rel = - (Z * alpha)**4 * m * c**2 / (2 * n**4) * (4/n - 3/(l + sp.Rational(1/2)))

#spin orbit correction
E_so = (Z * alpha)**4 * m * c**2 / (n**3) * (l / ((l + sp.Rational(1/2)) * (l + 1)))

#fine structure correction
E_fs = E_rel + E_so

#E in terms of j
E_final = E_fs.subs(j, l - sp.Rational(1/2))

#numerical function
Z_val = 5
alpha_val = 1/137
m_val = 5
c_val = 500
n_val = 2

#symbolic to expression 
E_rel_func = sp.lambdify(l, E_rel.subs({Z: Z_val, alpha: alpha_val, m: m_val, c: c_val, n: n_val}), 'numpy')
E_so_func = sp.lambdify(l, E_so.subs({Z: Z_val, alpha: alpha_val, m: m_val, c: c_val, n: n_val}), 'numpy')
E_fs_func = sp.lambdify(l, E_final.subs({Z: Z_val, alpha: alpha_val, m: m_val, c: c_val, n: n_val}), 'numpy')

#range of l
l_values = np.linspace(1, 10, 100)

#fig ax call
fig, ax = plt.subplots()
ax.set_xlim(-1, 10)
#ax.set_ylim(-1e-17, 1e-17)
ax.set_xlabel("Orbital Angular Momentum l")
ax.set_ylabel("Energy Correction (J)")
ax.set_title("Fine Structure Energy Corrections")


line_rel, = ax.plot([], [], 'r-', label='Relativistic Correction')
line_so, = ax.plot([], [], 'g-', label='Spin-Orbit Correction')
line_fs, = ax.plot([], [], 'b-', label='Fine Structure Correction')
ax.legend()

#animate
def update(frame):
    l_vals = l_values[:frame]
    line_rel.set_data(l_vals, E_rel_func(l_vals))
    line_so.set_data(l_vals, E_so_func(l_vals))
    line_fs.set_data(l_vals, E_fs_func(l_vals))
    return line_rel, line_so, line_fs

ani = animation.FuncAnimation(fig, update, frames=len(l_values), interval=50, blit=True)
ani.save('p2_energy_ani.gif',fps=20)
plt.show()
```
